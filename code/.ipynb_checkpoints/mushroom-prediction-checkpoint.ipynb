## Clean the categorical dataset 

Replace Na with missing and anywhere the count is less than 100 we classify as noise as its real small percentage of data

```python
def cleaner(df):
    for col in categorical_features:
        df[col] = df[col].fillna('missing')
        df.loc[df[col].value_counts(dropna=False)[df[col]].values < 100, col] = "noise"
        df[col] = df[col].astype('category')

    return df
```

```python
train_df = cleaner(train_df)
test_df = cleaner(test_df)
```

Replace Cap-Diamtere with the mean of the value for missing records.

```python
cap_diameter_mean = pd.concat([train_df['cap-diameter'], test_df['cap-diameter']]).mean(numeric_only=True)
train_df['cap-diameter'].fillna(cap_diameter_mean, inplace=True)
test_df['cap-diameter'].fillna(cap_diameter_mean, inplace=True)
```

```python
def map_predictions(pred):
    if pred == 0:
        return 'e'  # Edible
    else:
        return 'p'  # Poisonous

# Apply the mapping to the binary predictions
final_labels = [map_predictions(pred) for pred in y_pred]
if 'e' in final_labels:
    print("E exists")

# If you want to create a DataFrame to store the results
submission = pd.DataFrame({
    'id': test_df.index,  # Replace 'Id' with your identifier column name if needed
    'class': final_labels
})
submission.head()
output_data_file ="".join([data_folder,'/output/submission.csv'])
submission.to_csv(output_data_file, index=False)
```
